```python
# scripts/run_iforest_colab.py (versi√≥n corregida para Colab + Drive)

import os
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import plotly.express as px

# üß≠ Ruta correcta al proyecto en Drive
project_root = "/content/drive/MyDrive/TFG/ALIEN"

# ‚úÖ Rutas completas reconstruidas
train_pca_path = os.path.join(project_root, "results", "pca", "train", "pca_results.csv")
test_pca_path = os.path.join(project_root, "results", "pca", "test", "pca_results.csv")

train_dir = os.path.join(project_root, "data", "processed", "train")
test_dir = os.path.join(project_root, "data", "processed", "test")

result_dir = os.path.join(project_root, "results", "isolation_forest")
os.makedirs(result_dir, exist_ok=True)

# 1Ô∏è‚É£ Cargar resultados PCA
df_train = pd.read_csv(train_pca_path)
df_test = pd.read_csv(test_pca_path)

# 2Ô∏è‚É£ Asociar nombres (ya est√°n en las columnas, no hace falta renombrar)
# Solo aseguramos que existen
assert 'Star' in df_train.columns
assert 'Star' in df_test.columns

# 3Ô∏è‚É£ Entrenar modelo Isolation Forest
print("üöÄ Entrenando Isolation Forest...")
iso_forest = IsolationForest(contamination='auto', random_state=42)
iso_forest.fit(df_train[['PC1', 'PC2']])

# 4Ô∏è‚É£ Predecir anomal√≠as en test
print("üîç Detectando anomal√≠as en test...")
df_test['Anomaly'] = iso_forest.predict(df_test[['PC1', 'PC2']])
df_test['Score'] = iso_forest.decision_function(df_test[['PC1', 'PC2']])

# 5Ô∏è‚É£ Visualizaci√≥n
fig = px.scatter(
    df_test,
    x='PC1',
    y='PC2',
    color='Anomaly',
    hover_data=['Star', 'Score'],
    title='üîç Anomaly Detection using Isolation Forest',
    labels={'PC1': 'Principal Component 1', 'PC2': 'Principal Component 2', 'Anomaly': 'Anomaly'},
    template='plotly_white'
)
fig.update_traces(marker=dict(size=10))
fig.show()

# 6Ô∏è‚É£ Guardar resultados
df_test.to_csv(os.path.join(result_dir, "pca_isolation_forest_result.csv"), index=False)
fig.write_html(os.path.join(result_dir, "isolation_forest_plot.html"))

print("‚úÖ Resultados guardados en:", result_dir)

anomalies = df_test[df_test["Anomaly"] == -1]
print("üå† Estrellas detectadas como an√≥malas:")
print(anomalies[["Star", "PC1", "PC2", "Score"]])

```

```python
# scripts/run_autoencoder.py

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.pca_analysis import load_processed_data
from src.autoencoder_model import (
    normalize_data,
    build_autoencoder,
    train_autoencoder,
    compute_reconstruction_errors,
    detect_anomalies,
    plot_reconstruction_error,
    plot_errors_per_star
)

# üìÅ Rutas
project_root = "/content/drive/MyDrive/TFG/ALIEN"
train_dir = os.path.join(project_root, "data", "processed", "train")
test_dir = os.path.join(project_root, "data", "processed", "test")
result_dir = os.path.join(project_root, "results", "autoencoder")
os.makedirs(result_dir, exist_ok=True)

# 1Ô∏è‚É£ Cargar datos de entrenamiento
print("üì• Cargando curvas de entrenamiento...")
train_matrix, train_names = load_processed_data(train_dir)
train_matrix_norm, scaler = normalize_data(train_matrix)

# 2Ô∏è‚É£ Crear y entrenar el autoencoder
print("üß† Entrenando Autoencoder...")
autoencoder = build_autoencoder(input_dim=train_matrix_norm.shape[1])
train_autoencoder(autoencoder, train_matrix_norm, epochs=50, batch_size=8)

# 3Ô∏è‚É£ Errores de reconstrucci√≥n sobre train
print("üìà Calculando errores de reconstrucci√≥n (entrenamiento)...")
train_errors = compute_reconstruction_errors(autoencoder, train_matrix_norm)
train_threshold = np.percentile(train_errors, 90)
train_labels = detect_anomalies(train_errors, train_threshold)

# 4Ô∏è‚É£ Guardar resultados entrenamiento
df_train = pd.DataFrame({
    'Star': train_names,
    'ErrorReconstruccion': train_errors,
    'Anomalia': train_labels
})
df_train.to_csv(os.path.join(result_dir, 'autoencoder_train_results.csv'), index=False)

# 5Ô∏è‚É£ Visualizaci√≥n entrenamiento
plot_reconstruction_error(train_errors, train_threshold)
plot_errors_per_star(train_errors, train_threshold)

# 6Ô∏è‚É£ Cargar y normalizar datos de test
print("üì• Cargando curvas de test...")
test_matrix, test_names = load_processed_data(test_dir)
test_matrix_norm = scaler.transform(test_matrix)

# 7Ô∏è‚É£ Errores de reconstrucci√≥n sobre test
print("üìà Calculando errores de reconstrucci√≥n (test)...")
test_errors = compute_reconstruction_errors(autoencoder, test_matrix_norm)
test_labels = detect_anomalies(test_errors, train_threshold)  # umbral de train

# 8Ô∏è‚É£ Guardar resultados test
df_test = pd.DataFrame({
    'Star': test_names,
    'ErrorReconstruccion': test_errors,
    'Anomalia': test_labels
})
df_test.to_csv(os.path.join(result_dir, 'autoencoder_test_results.csv'), index=False)

# 9Ô∏è‚É£ Visualizaci√≥n test
plot_reconstruction_error(test_errors, train_threshold)
plot_errors_per_star(test_errors, train_threshold)

print("‚úÖ An√°lisis con Autoencoder completado.")
```

```python
# ‚úÖ test_isolation_forest.py ‚Äì test funcional con visualizaci√≥n

import os
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import plotly.express as px

# Ruta del proyecto
project_root = "/content/drive/MyDrive/TFG/ALIEN"

# Rutas de entrada
train_pca_file = os.path.join(project_root, "results", "pca", "train", "pca_results.csv")
test_pca_file  = os.path.join(project_root, "results", "pca", "test",  "pca_results.csv")

# Directorio de resultados
output_dir = os.path.join(project_root, "results", "isolation_forest")
os.makedirs(output_dir, exist_ok=True)

# Cargar datos PCA
df_train = pd.read_csv(train_pca_file)
df_test = pd.read_csv(test_pca_file)

assert 'PC1' in df_train.columns and 'PC2' in df_train.columns, "‚ùå PCA mal cargado"
assert 'Star' in df_test.columns, "‚ùå Falta columna 'Star'"

# Entrenar modelo
print("üöÄ Entrenando Isolation Forest...")
model = IsolationForest(contamination='auto', random_state=42)
model.fit(df_train[['PC1', 'PC2']])

# Predecir sobre test
df_test['Anomaly'] = model.predict(df_test[['PC1', 'PC2']])  # -1 = an√≥mala
df_test['Score'] = model.decision_function(df_test[['PC1', 'PC2']])

# Cu√°ntas an√≥malas
n_anomalies = (df_test['Anomaly'] == -1).sum()
print(f"üîé Anomal√≠as detectadas: {n_anomalies} de {len(df_test)} curvas")

# Gr√°fico interactivo
fig = px.scatter(
    df_test,
    x='PC1', y='PC2',
    color='Anomaly',
    title='Detecci√≥n de anomal√≠as (Isolation Forest)',
    hover_data=['Star', 'Score'],
    template='plotly_white'
)
fig.update_traces(marker=dict(size=9))
fig.show()

# Guardar
df_test.to_csv(os.path.join(output_dir, "iforest_test_results.csv"), index=False)
fig.write_html(os.path.join(output_dir, "iforest_test_plot.html"))

print("‚úÖ Resultados guardados.")
```

```python
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# 1. Crear datos sint√©ticos
def create_synthetic_data(n_normal=50, n_anomalous=5, length=100):
    x = np.linspace(0, 4 * np.pi, length)

    # Normales: senos y cosenos
    normals = [np.sin(x + np.random.uniform(-0.1, 0.1)) for _ in range(n_normal // 2)]
    normals += [np.cos(x + np.random.uniform(-0.1, 0.1)) for _ in range(n_normal // 2)]

    # An√≥malos: ruido, picos, constantes
    anomalies = [np.random.normal(0, 1, length) for _ in range(n_anomalous // 3)]
    anomalies += [np.ones(length) * np.random.uniform(0.5, 1.5) for _ in range(n_anomalous // 3)]
    anomalies += [np.concatenate([np.sin(x[:50]), np.ones(50)]) for _ in range(n_anomalous // 3)]

    data = np.array(normals + anomalies)
    labels = np.array([1] * len(normals) + [-1] * len(anomalies))
    return data, labels

# 2. Normalizar
def normalize(data):
    scaler = MinMaxScaler()
    return scaler.fit_transform(data.T).T, scaler

# 3. Autoencoder simple
def build_autoencoder(input_dim):
    inp = Input(shape=(input_dim,))
    x = Dense(64, activation='relu')(inp)
    x = Dense(32, activation='relu')(x)
    x = Dense(64, activation='relu')(x)
    out = Dense(input_dim, activation='sigmoid')(x)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer=Adam(0.001), loss='mse')
    return model

# 4. Entrenamiento y evaluaci√≥n
X, y_true = create_synthetic_data()
X_norm, _ = normalize(X)

X_train = X_norm[y_true == 1]  # solo normales

model = build_autoencoder(X.shape[1])
model.fit(X_train, X_train, epochs=50, batch_size=8, validation_split=0.1, callbacks=[EarlyStopping(patience=5)], verbose=0)

# Errores de reconstrucci√≥n
reconstructed = model.predict(X_norm)
errors = np.mean(np.square(X_norm - reconstructed), axis=1)

# Umbral simple: 95 percentil de errores normales
threshold = np.percentile(errors[y_true == 1], 98)

y_pred = np.where(errors > threshold, -1, 1)

# Verificaci√≥n
print("Errores:", errors)
print("Umbral:", threshold)
print("Anomal√≠as detectadas:", np.sum((y_pred == -1) & (y_true == -1)))
print("Falsos positivos:", np.sum((y_pred == -1) & (y_true == 1)))

assert np.sum((y_pred == -1) & (y_true == -1)) >= len(y_true[y_true == -1]) - 1, "‚ùå No detect√≥ todas las anomal√≠as"
assert np.sum((y_pred == -1) & (y_true == 1)) <= 1, "‚ùå Falsos positivos altos"

print("‚úÖ Test sint√©tico del Autoencoder pasado correctamente üéâ")
```
